```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import pickle

# Load the combo list
file_path = r"F:\combo_lists\whoi_combo_list.pickle"
with open(file_path, 'rb') as file:
    combo_list = pickle.load(file)

# Function to load and display images
def load_and_display_images(directories, num_images=5, size=64):
    plt.figure(figsize=(10, 10))
    for i, directory in enumerate(directories, 1):
        image_files = os.listdir(directory)
        num_images = min(num_images, len(image_files))
        for j in range(num_images):
            image_path = os.path.join(directory, image_files[j])
            image = Image.open(image_path)
            image = image.resize((size, size), Image.ANTIALIAS)
            plt.subplot(len(directories), num_images, (i-1)*num_images + j + 1)
            plt.imshow(image)
            plt.axis('off')
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.show()

# Directories to display images from
directories = combo_list[70:80]

# Display all images together
load_and_display_images(directories, size=64)
```
