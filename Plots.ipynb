```python
import matplotlib.pyplot as plt

# Define x-axis data (annotation budget)
x = [10, 40, 80]

# Define y-axis data (accuracy) for both lines
y1 = [ 32.5 , 37.41, 42.3]

y2 = [ 20.5 , 29.7, 34.5]

# Create the plot
fig, ax = plt.subplots()

# Plot the lines with different colors and line styles
ax.plot(x, y1, label="Our Method", color="blue", linestyle="-", linewidth=2, marker='o', markersize=8)
ax.plot(x, y2, label="RAND", color="orange", linestyle="--", linewidth=2, marker='s', markersize=8)

# Set labels for x-axis, y-axis, and legend
ax.set_xlabel("Annotation Budget", fontsize=12)
ax.set_ylabel("Accuracy", fontsize=12)
ax.legend(fontsize=10)

# Set grid and background color
ax.grid(True)
ax.set_facecolor('whitesmoke')

# Set title
ax.set_title(" CIFAR100 results with variable AL annotation Budget : Accuracy vs. Budget", fontsize=14)

# Display the plot
plt.show()
```
