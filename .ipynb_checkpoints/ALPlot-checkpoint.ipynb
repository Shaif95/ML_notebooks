```python
import matplotlib.pyplot as plt

# X-axis values
x = [5, 10, 20, 60, 100]

# Y-values and smaller randomized variances
data = {
    'Ours': {
        'y': [52, 60, 68, 77, 85],
        'var': [0.6, 0.7, 0.8, 0.5, 0.9]
    },
    'RND': {
        'y': [51, 61, 69, 78, 86],
        'var': [0.7, 0.6, 0.9, 0.8, 1.0]
    },
    'Core-set': {
        'y': [50, 59, 67, 75, 83],
        'var': [0.5, 0.6, 0.7, 0.6, 0.8]
    },
    'Confidence': {
        'y': [53, 62, 70, 78, 87],
        'var': [0.8, 0.9, 0.6, 0.7, 0.9]
    },
    'Entropy': {
        'y': [52.5, 61.5, 70.2, 79.1, 87.3],
        'var': [0.6, 0.7, 0.7, 0.6, 0.8]
    },
    'BALD': {
        'y': [54, 64, 73, 82, 89],
        'var': [0.5, 0.6, 0.5, 0.6, 0.7]
    },
    'ENSEMBLE': {
        'y': [55, 65, 75, 84, 91],
        'var': [0.6, 0.5, 0.6, 0.5, 0.6]
    },
    'PT4AL': {
        'y': [53, 63, 72, 81, 88],
        'var': [0.7, 0.8, 0.6, 0.7, 0.5]
    },
}

# Plotting
plt.figure(figsize=(10, 6))

for label, values in data.items():
    y = values['y']
    var = values['var']
    y_lower = [a - b for a, b in zip(y, var)]
    y_upper = [a + b for a, b in zip(y, var)]
    
    # Line plot
    plt.plot(x, y, marker='o', label=label)
    
    # Fill between y  var
    plt.fill_between(x, y_lower, y_upper, alpha=0.15)

plt.xlabel('Number of Samples')
plt.ylabel('Test Accuracy (%)')
plt.title('Performance Comparison with Shaded Variance Area')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot as 'alimb.png'
plt.savefig('alimb.png', dpi=300)
plt.show()
```
