```python
import dask.dataframe as dd
import numpy as np
from tqdm import tqdm

# Define the paths to the CSV files
file_paths = [
    r"E:\imagenet\lbpfeature1.csv",
    r"E:\imagenet\lbpfeature2.csv",
    r"E:\imagenet\lbpfeature3.csv",
    r"E:\imagenet\lbpfeature4.csv"
]

# Load the CSV files into Dask DataFrames and immediately drop rows with NaN values
dfs = [dd.read_csv(file_path).dropna() for file_path in file_paths]

# Function to get random indexes and retrieve the full rows
def get_random_rows(df, n=250000):
    # Convert the Dask DataFrame to a Pandas DataFrame after dropping NaNs
    pandas_df = df.compute()
    
    # Get the total number of rows in the DataFrame
    total_rows = len(pandas_df)
    
    # If there are fewer rows than n, use the total number of rows instead
    n = min(total_rows, n)
    
    # Generate random indexes
    random_indexes = np.random.choice(total_rows, n, replace=False)
    
    # Retrieve the full rows for these random indexes and split on tabs
    random_rows = [
        [item for sublist in (str(cell).split('\t') for cell in row) for item in sublist]
        for row in pandas_df.iloc[random_indexes].values
    ]
    
    return random_rows

# Initialize an empty list to store all rows
all_random_rows = []

# Get the rows from each DataFrame
for df in tqdm(dfs, desc="Processing DataFrames"):
    rows = get_random_rows(df)
    all_random_rows.extend(rows)

# Print the total number of rows collected
print(len(all_random_rows))
```

```python
# Remove the first and last items from each element in all_random_rows
new_all_random_rows = [row[1:-1] for row in all_random_rows]
import numpy as np
# Assuming new_all_random_rows is a list of lists (rows)
new_all_random_rows = np.array(new_all_random_rows)
# Convert the dataset to a numeric type if it's not already
new_all_random_rows = new_all_random_rows.astype(np.float64)
print(np.shape(new_all_random_rows))  # Confirm the shape of the array
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ensure data is numeric
new_all_random_rows = new_all_random_rows.astype(np.float64)

# Calculate mean and variance of features
means = np.mean(new_all_random_rows, axis=0)
variances = np.var(new_all_random_rows, axis=0)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(range(len(means)), means, color='skyblue', alpha=0.7, label='Mean')
plt.plot(range(len(variances)), variances, color='orange', linewidth=2, label='Variance')
plt.title('Average Feature Values with Variance')
plt.xlabel('Feature Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(new_all_random_rows)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c='purple', alpha=0.5)
plt.title('PCA 2D Projection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Fit PCA for variance explanation
pca = PCA().fit(new_all_random_rows)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title('Variance Explained by Principal Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
import seaborn as sns

# Calculate correlation matrix
correlation_matrix = np.corrcoef(new_all_random_rows, rowvar=False)

# Plotting
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Feature Correlation')
plt.show()
```

```python
import pandas as pd
from tqdm import tqdm

# Define the paths to the CSV files
file_paths = [
    r"E:\imagenet\lbpfeature1.csv",
    r"E:\imagenet\lbpfeature2.csv",
    r"E:\imagenet\lbpfeature3.csv",
    r"E:\imagenet\lbpfeature4.csv"
]

# Load and combine all CSV files into a single DataFrame with a progress bar
all_dataframes = []
for file_path in tqdm(file_paths, desc="Loading CSV files"):
    df = pd.read_csv(file_path, header=None)  # Read the CSV without a header
    df = df.iloc[1:]  # Remove the first row
    all_dataframes.append(df)

combined_df = pd.concat(all_dataframes, ignore_index=True)

# Convert the combined DataFrame to a list of lists (rows)
all_random_rows = combined_df.values.tolist()

# Split the tab-separated strings in each row into individual elements
split_rows = [
    [item for sublist in (str(cell).split('\t') for cell in row) for item in sublist]
    for row in all_random_rows
]

# Remove the first and last items from each row
new_all_random_rows = [row[1:-1] for row in split_rows]
import numpy as np

# Convert the list of lists to a NumPy array
new_all_random_rows = np.array(new_all_random_rows)

# Convert the dataset to a numeric type if it's not already
new_all_random_rows = new_all_random_rows.astype(np.float64)

# Confirm the shape of the array
print("Final shape of new_all_random_rows:", np.shape(new_all_random_rows))
```

```python
import numpy as np
from sklearn.cluster import MiniBatchKMeans
import matplotlib.pyplot as plt

# Apply MiniBatchKMeans with 1000 clusters
n_clusters = 1000
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=1024, random_state=42)
kmeans.fit(new_all_random_rows)

# Get the cluster labels for each point
cluster_labels = kmeans.labels_

# Count the number of elements in each cluster
cluster_counts = np.bincount(cluster_labels, minlength=n_clusters)

# Plot the bar chart for the cluster sizes
plt.figure(figsize=(12, 6))
plt.bar(range(n_clusters), cluster_counts, color='blue')
plt.xlabel('Cluster ID')
plt.ylabel('Number of Elements')
plt.title('Number of Elements in Each Cluster')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt

# Apply Spectral Clustering with 1000 clusters
n_clusters = 1000
spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42, n_jobs=-1)
cluster_labels = spectral_clustering.fit_predict(new_all_random_rows)

# Count the number of elements in each cluster
cluster_counts = np.bincount(cluster_labels, minlength=n_clusters)

# Plot the bar chart for the cluster sizes
plt.figure(figsize=(12, 6))
plt.bar(range(n_clusters), cluster_counts, color='blue')
plt.xlabel('Cluster ID')
plt.ylabel('Number of Elements')
plt.title('Number of Elements in Each Cluster (Spectral Clustering)')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt

# Apply Agglomerative Clustering
n_clusters = 1000  # Number of clusters to form
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')
cluster_labels = agg_clustering.fit_predict(new_all_random_rows)

# Count the number of elements in each cluster
unique_labels, cluster_counts = np.unique(cluster_labels, return_counts=True)

# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(unique_labels, cluster_counts, color='blue')
plt.xlabel('Cluster ID')
plt.ylabel('Number of Elements')
plt.title('Number of Elements in Each Cluster (Agglomerative Clustering)')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5, n_jobs=-1)  # Adjust eps and min_samples as needed
dbscan.fit(new_all_random_rows)

# Get cluster labels
cluster_labels = dbscan.labels_

# Count the number of elements in each cluster
# DBSCAN labels noise as -1, so we include it in the counts
unique_labels, cluster_counts = np.unique(cluster_labels, return_counts=True)

# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(unique_labels, cluster_counts, color='blue')
plt.xlabel('Cluster ID')
plt.ylabel('Number of Elements')
plt.title('Number of Elements in Each Cluster (DBSCAN)')
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Set the directory path
folder_path = r'G:\datasets\Underwater_Image\WHOI\archive\dataset_pm\training'

# Dictionary to hold subfolder names and their respective file counts
folder_counts = {}

# Walk through each subfolder in the directory
for root, dirs, files in os.walk(folder_path):
    # Ignore the main folder, only count subfolders
    if root != folder_path:
        folder_name = os.path.basename(root)
        folder_counts[folder_name] = len(files)

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(folder_counts.keys(), folder_counts.values())
plt.xticks(rotation=90)
plt.xlabel('Subfolder Names')
plt.ylabel('Number of Files')
plt.title('Number of Elements in Each Subfolder')
plt.tight_layout()
plt.show()
```
