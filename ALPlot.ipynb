```python
import matplotlib.pyplot as plt

# X-axis values
x = [10, 20, 40]

# Y-values and variances
data = {
    'Ours': {
        'y': [50.0, 56.73, 65.2],
        'var': [0.6, 0.7, 1.8]
    },
    'Contrastive(RND)': {
        'y': [45.7, 55.6, 64.9],
        'var': [0.7, 0.6, 1.1]
    },
    'Core-set': {
        'y': [43.45, 53.2, 63.9],
        'var': [0.5, 0.6, 1.3]
    },
    'Confidence': {
        'y': [22.1, 29.4, 49.3],
        'var': [0.8, 0.7, 1.1]
    },
    'Entropy': {
        'y': [31.2, 33.7, 54.9],
        'var': [0.5, 0.6, 1.7]
    },
    'BALD': {
        'y': [44.4, 52.6, 60.2],
        'var': [0.5, 0.6, 1.7]
    },
    'ENSEMBLE': {
        'y': [40.45, 48.6, 59.45],
        'var': [0.6, 0.5, 1.6]
    },
    'PT4AL': {
        'y': [47.2, 55.3, 65.5],
        'var': [0.7, 0.7, 0.5]
    },
    'Sup+KMeans': {
        'y': [42.5, 47.0, 57.8],
        'var': [0.6, 0.5, 1.4]
    },
}

# Visually distinct, non-green colors
colors = [
    "black",       # Ours
    "gold",        # Contrastive(RND)
    "red",         # Core-set
    "orange",      # Confidence
    "blue",        # Entropy
    "gray",        # BALD
    "skyblue",     # ENSEMBLE
    "indigo",      # PT4AL
    "magenta"      # Sup+KMeans
]

# Plotting
plt.figure(figsize=(10, 6))

for i, (label, values) in enumerate(data.items()):
    y = values['y']
    var = values['var']
    y_lower = [a - b for a, b in zip(y, var)]
    y_upper = [a + b for a, b in zip(y, var)]

    if label == 'Ours':
        plt.plot(x, y, marker='o', label=label, color=colors[i], linewidth=3, markersize=8)
        plt.fill_between(x, y_lower, y_upper, alpha=0.25, color=colors[i])
    else:
        plt.plot(x, y, marker='o', label=label, color=colors[i], linewidth=1.8)
        plt.fill_between(x, y_lower, y_upper, alpha=0.15, color=colors[i])

plt.xlabel('Percent of Initial Samples selected')
plt.ylabel('Test Accuracy (%)')
plt.title('Performance Comparison with Shaded Variance Area')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('alimb.png', dpi=300)
plt.show()
```
