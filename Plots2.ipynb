```python
import matplotlib.pyplot as plt

methods = [
    "Att-LSTM-S (Linear)",
    "Att-LSTM-S (2D CNN)",
    "Att-LSTM-B (Linear)",
    "Att-LSTM-B (2D CNN)",

    "ViVIT",
    "LRCN",
    "VGG CNN"
]

f1_scores = [
    99.18, 99.15, 98.72, 98.56, 92.39,  89.26, 90.8
]

plt.figure(figsize=(15, 4))
plt.bar(methods, f1_scores, color='skyblue')
plt.xlabel('F1 Score')
plt.ylabel('Method')
plt.title('F1 Score for Different Methods')
plt.gca()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define your data
models = ["ResNet(32,32)", "Ours(21k)", "rnd1k"]
accuracy = [73.62, 73.94, 70.6]

# Create a bar chart
plt.figure(figsize=(8, 6))

# Plot bars
plt.bar(models, accuracy, color=['blue', 'green', 'red'])

# Add labels and a title
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison for Different Models (Cifar10)')

# Show the bar chart
plt.show()
```
